```python
import pandas as pd
from math import log
from ast import literal_eval
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
```

```python
df = pd.read_csv('/content/drive/MyDrive/munning_rachine/dataset/keyword/PlaystoreReviewData_keyword.csv', converters={"data": literal_eval, "label":literal_eval})
works = pd.read_csv('/content/drive/MyDrive/munning_rachine/dataset/keyword/PlaystoreReviewData_top_keyword.csv', encoding='euc-kr', index_col=False)
tag = list(set(list(works.iloc[:,0])+list(works.iloc[:,2])))
```

```python
tag = tag[1:]
tag
```

```python
df_n = df['Comment'].apply(lambda x: np.str_(x))
```

```python
df_n
```

```python
tdflist = []
t = []
d = {}
vectorizer = TfidfVectorizer()
sp_matrix = vectorizer.fit_transform(df_n)

word2id = defaultdict(lambda : 0)
for idx, feature in enumerate(vectorizer.get_feature_names()):
    word2id[feature] = idx
    
for i, sent in enumerate(df_n):    
    for token in sent.split(','):
      if sp_matrix[i, word2id[token]] == 0:
        pass
      elif token in tag and token not in t:
            t.append(token)
      else:
        d[token] = sp_matrix[i, word2id[token]] #딕셔너리에 숫자값을 넣음    
    d = sorted(d, key= lambda x : d[x], reverse=True)
    t = list(set(list(d[:3])+t))
    tdflist.append(t)
    d = {} #초기화
    t = []
```

```python
tdflist
```

```python
len(tdflist)
```

```python
col = []
temp = []
for i in range(len(tdflist)):
  temp.append(len(tdflist[i]))
  max_num = max(temp)
  
for i in range(max_num):
  col.append('Keyword' + str(i+1))
```

```python
dd = pd.DataFrame(tdflist, columns=col)
dd
```

```python
#맨 앞 $기호 제거
for i in range(len(df['Comment_processed'])):
  df['Comment_processed'][i] = df['Comment_processed'][i][1:]
```

```python
split = df.Comment_processed.str.split('$')
split = split.apply(lambda x: pd.Series(x))
split.stack().reset_index(level=1, drop=True)
split = split.stack().reset_index(level=1, drop=True).to_frame('split')
split
```

```python
df = df.merge(split, left_index=True, right_index=True, how='left')
```

```python
def find_sentence(keyword_dict, dataframe):
  for dict_keyname in keyword_dict:
    dataframe[dict_keyname] = 0
    for (i, row) in enumerate(dataframe['split']):
        for keyword in keyword_dict[dict_keyname]:
          if keyword in row:
            dataframe[dict_keyname].iloc[i] = 1
  return dataframe
```

```python
# 딕셔너리와 데이터 프레임 넣기
find_sentence(keyword_dict, df)
```
